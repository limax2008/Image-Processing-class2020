```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# import skimage
# plt.rcParams['figure.figsize']=(8.0,6.0)
```

# Erosion

```python
im = cv2.imread('chessboard.jpg',0)

# kernel
kernel = np.ones((5,5),np.uint8)
ime = cv2.erode(im, kernel, iterations=1)

plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.subplot(122)
plt.imshow(ime, cmap='gray')
```

## Remove the noise

```python
im = np.zeros((100,100))
# set the noise
im[10:20, 20:30] =1
im [50:55,80:85]=1
print('original area:{}'.format(np.sum(im)))
ime = cv2.erode(im, kernel, iterations=3)
print('erode area:{}'.format(np.sum(ime)))

plt.subplot(121)
plt.imshow(im)
plt.subplot(122)
plt.imshow(ime)
```

# Dilation

```python
im = cv2.imread('chessboard.jpg',0)

# kernel
kernel = np.ones((5,5),np.uint8)
ime = cv2.dilate(im, kernel, iterations=1)

plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.subplot(122)
plt.imshow(ime, cmap='gray')
```

# Opening

```python
from skimage.util import random_noise
imNoise = random_noise(im, mode ='salt', seed=1)
imo = cv2.morphologyEx(imNoise, cv2.MORPH_OPEN, kernel)

plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.title('original image')
plt.subplot(132)
plt.imshow(imNoise, cmap='gray')
plt.title('noise image')

plt.subplot(133)
plt.imshow(imo, cmap='gray')
plt.title('opening image')
```

# Closing

```python
from skimage.util import random_noise
imNoise = random_noise(im, mode ='pepper', seed=1)
imo = cv2.morphologyEx(imNoise, cv2.MORPH_CLOSE, kernel)

plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.title('original image')
plt.subplot(132)
plt.imshow(imNoise, cmap='gray')
plt.title('noise image')

plt.subplot(133)
plt.imshow(imo, cmap='gray')
plt.title('closing image')
```

# Gradient

```python
img = cv2.morphologyEx(im, cv2.MORPH_GRADIENT, kernel)
plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.title('original image')
plt.subplot(122)
plt.imshow(img, cmap='gray')
plt.title('gradient image')
```

# Top hat

```python
# numpy to define the kernel
kernel = np.ones((5,5), np.uint8)
print('----numpy kernel:{}----'.format(kernel))
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
print('----opencv kernel:{}------'.format(kernel))
```

```python
imt = cv2.morphologyEx(im, cv2.MORPH_TOPHAT, kernel)
im0 = cv2.morphologyEx(im, cv2.MORPH_OPEN, kernel)
plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.title('original image')
plt.subplot(132)
plt.imshow(im0, cmap='gray')
plt.title('openinig image')
plt.subplot(133)
plt.imshow(imt, cmap='gray')
plt.title('topHat image')
```

#  Black Hat

```python
imt = cv2.morphologyEx(im, cv2.MORPH_BLACKHAT, kernel)
im0 = cv2.morphologyEx(im, cv2.MORPH_CLOSE, kernel)
plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.title('original image')
plt.subplot(132)
plt.imshow(im0, cmap='gray')
plt.title('closing image')
plt.subplot(133)
plt.imshow(imt, cmap='gray')
plt.title('blackHat image')
```
