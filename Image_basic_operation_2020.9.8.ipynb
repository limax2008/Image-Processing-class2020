## Image fill (extension)

Fill pixels around the image, just as the photo frame. This is often used in the convolution operation in deep learning.

```python
cv2.copyMakeBorder??
```

```python
BLUE=[255,0,0]
img1 = cv2.imread('opencv_logo.jpg',cv2.COLOR_BGR2GRAY)
print(img1.shape)
# img1=cv2.imread('car.png', cv2.COLOR_)
replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
print(replicate.shape)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)


plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL'), plt.axis('off')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE'),plt.axis('off')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT'),plt.axis('off')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101'),plt.axis('off')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP'),plt.axis('off')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT'),plt.axis('off')
```

```python
# load the image
img1 = cv2.imread('lena.png')
img2 = cv2.imread('opencv_logo.jpg')
img2 = cv2.resize(img2, (100,100))
 
# I want to put logo on top-left corner, So I create a ROI
rows,cols,channels = img2.shape 
roi = img1[0:rows, 0:cols ]
 
# Now create a mask of logo and create its inverse mask also
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

# binarize the image
ret, mask = cv2.threshold(img2gray, 175, 255, cv2.THRESH_BINARY)
print(mask)
# plt.imshow(mask)
mask_inv = cv2.bitwise_not(mask)
 
# Now black-out the area of logo in ROI
# roi mask,0
# mask=maskmask=mask_inv,'mask='
img1_bg = cv2.bitwise_and(roi,roi,mask = mask)
#   roi   mask_inv ,0
# Take only region of logo from logo image.
img2_fg = cv2.bitwise_and(img2,img2,mask = mask_inv)
 
# Put logo in ROI and modify the main image
dst = cv2.add(img1_bg,img2_fg) 
img1[0:rows, 0:cols ] = dst
 
cv2.imshow('res',img1) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
