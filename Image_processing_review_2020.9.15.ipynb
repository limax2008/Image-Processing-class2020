# Image definition

## What is an image

A two-dimensional function, f(x,y),x and y are spatial coordinates. The amplitude of f is called intensity or gray level at point (x,y).
In the computer, image can be represented as a matrix.

Color image can be represented as multichannels (R, G, B).

## What is image processing

Process digital images by means of computer; it covers low-, mid-, and high-level processes:
 1. low-level: inputs and outputs are images
 2. mid-level: outputs are attributes extracted from input images
 3. high-level: use computer to emulate human vision

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## read one color image and display it

```python
im = cv2.imread('lena.png')
# print(type(im))
# im = im[:,:,::-1]
im = im[...,::-1]
# print(type(im))
plt.imshow(im)
```

## split the color image into r g b channels

```python
r,g,b = cv2.split(im)
plt.subplot(1,3,1)
plt.imshow(r,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(g,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(b,cmap='gray')

# cv2.imshow('r', r)
# cv2.imshow('g', g)
# cv2.imshow('b', b)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
im1 = cv2.imread('lena.png')
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)

print('datatype:{}'.format(im1.dtype))
cnt_x = np.int(im1.shape[0]/2)
cnt_y = np.int(im1.shape[1]/2)
cv2.rectangle(np.array(im), (0,0),(50,50), (0, 0, 255),5)
# imd = im
roi_pt1=(cnt_x-50,cnt_y-50)
roi_pt2 = (cnt_x+50, cnt_y+50)

print('pt1:{},pt2:{}'.format(roi_pt1, roi_pt2))
print('image dimension:{}'.format(im1.shape))

cv2.rectangle((im1), roi_pt1,roi_pt2,(255,0,0),5)
plt.subplot(1,2,1)
plt.imshow(im1)
# cv2.imshow('image',im)


roi = im1[cnt_x-50:cnt_x+50,cnt_y-50:cnt_y+50]
plt.subplot(1,2,2)
plt.imshow(roi)
# cv2.imshow('roi',roi)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

## draw line R channel

```python
r3 = cv2.merge((r,r,r))
cv2.line(r3, (0,0), (200,200), (0,0.,255), 5)
plt.imshow(r3)
```

## draw circle on G channel

```python
g3 = cv2.merge((g,g,g))
cv2.circle(g3, (100,100), 50, (255,0.,0), 5)
plt.imshow(g3)
```

## merge the three channels

```python
r3 = cv2.cvtColor(r3, cv2.COLOR_RGB2GRAY)
g3 = cv2.cvtColor(g3, cv2.COLOR_RGB2GRAY)
b3 = cv2.cvtColor(b3, cv2.COLOR_RGB2GRAY)
im = cv2.merge((r3,g3,b3))
plt.imshow(r3)
# cv2.imshow('merge image',im)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
